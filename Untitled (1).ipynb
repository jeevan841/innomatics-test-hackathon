```python
import numpy as nu
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import mysql.connector
```

```python
df_csv = pd.read_csv('orders.csv')
```

```python
df_json = pd.read_json('users.json')
```

```python
db = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "Mysql@bond@1",
    database = "new_schema1"
)
mycursor = db.cursor()
print(db)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
query = "SELECT * FROM restaurants"
df_sql = pd.read_sql(query, db)
print(df_sql.head())
print(df_sql.info())
```

```python
df_csv = pd.DataFrame(df_csv)
print(df_csv)
```

```python
df_json = pd.DataFrame(df_json)
print(df_json)
```

```python
df_csv['user_id'] = df_csv['user_id'].astype(int)
df_json['user_id'] = df_json['user_id'].astype(int)
df_csv['restaurant_id'] = df_csv['restaurant_id'].astype(int)
df_sql['restaurant_id'] = df_sql['restaurant_id'].astype(int)
```

```python
#merging the data files
df_first_merge = pd.merge(df_csv, df_sql, on='restaurant_id', how='left', suffixes=('', '_rest'))
df_final_merge = pd.merge(df_first_merge, df_json, on='user_id', how='left', suffixes=('', '_user'))
```

```python
df = pd.DataFrame(df_final_merge)
df.to_csv('final_food_dataset.csv', index=False)
```

```python
df_final_merge['order_date'] = pd.to_datetime(df_final_merge['order_date'], format='%d-%m-%Y')
df_final_merge['month'] = df_final_merge['order_date'].dt.to_period('M').astype(str)
```

```python
df_final_merge.to_csv('final_food_delivery_dataset.csv', index=False)
print("Final dataset created successfully!")
```

```python
plt.figure(figsize=(12, 5))
monthly_orders = df_final_merge.groupby('month').size()
monthly_orders.plot(kind='line', marker='o', color='#1f77b4', linewidth=2)
plt.title('Monthly Order Trends', fontsize=15)
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
city_revenue = df_final_merge.groupby('city')['total_amount'].sum().sort_values(ascending=False)
sb.barplot(x=city_revenue.index, y=city_revenue.values, palette='viridis')
plt.title('Total Revenue by City', fontsize=15)
plt.ylabel('Revenue ($)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
cuisine_revenue = df_final_merge.groupby('cuisine')['total_amount'].sum().sort_values(ascending=False)
sb.barplot(x=cuisine_revenue.index, y=cuisine_revenue.values, palette='magma')
plt.title('Total Revenue by Cuisine Type', fontsize=15)
plt.ylabel('Revenue ($)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.kdeplot(data=df_final_merge, x='total_amount', hue='membership', fill=True, palette='Set1')
plt.title('Spending Distribution: Gold vs Regular Members', fontsize=15)
plt.xlabel('Order Amount ($)')
plt.show()
```

```python
user_spending = df_final_merge.groupby('user_id')['total_amount'].sum()
top_spenders = user_spending[user_spending > 1000]
distinct_user_count = len(top_spenders)
print(f"Number of distinct users with total spending > 1000: {distinct_user_count}")
```

```python
bins = [0, 1, 2, 3, 4, 5]
labels = ['0-1', '1-2', '2-3', '3-4', '4-5']
df_final_merge['rating_range'] = pd.cut(df_final_merge['rating'], bins=bins, labels=labels)
revenue_by_rating = df_final_merge.groupby('rating_range')['total_amount'].sum()
print(revenue_by_rating.sort_values(ascending=False))
```

```python
gold_data = df_final_merge[df_final_merge['membership'] == 'Gold']
city_av = gold_data.groupby('city')['total_amount'].mean()
print(city_aov.sort_values(ascending=False))
```

```python
cuisine_analysis = df_final_merge.groupby('cuisine').agg(
    distinct_restaurants=('restaurant_id', 'nunique'),
    total_revenue=('total_amount', 'sum')
)
print(cuisine_analysis.sort_values(by='distinct_restaurants'))
```

```python
total_orders = len(df_final_merge)
gold_orders = len(df_final_merge[df_final_merge['membership'] == 'Gold'])
percentage = round((gold_orders / total_orders) * 100)
print(f"Percentage of orders by Gold members: {percentage}%")
```

```python
restaurant_stats = df_final_merge.groupby(['restaurant_id', 'restaurant_name']).agg(
    order_count=('order_id', 'count'),
    avg_order_value=('total_amount', 'mean')
).reset_index()
niche = restaurant_stats[restaurant_stats['order_count'] < 20]
top_niche = niche.sort_values(by='avg_order_value', ascending=False).head(1)
print(top_niche)
```

```python
membership_cuisine_matrix = df_final_merge.groupby(['membership', 'cuisine'])['total_amount'].sum().unstack()
print("--- Revenue Matrix: Membership vs. Cuisine ---")
print(membership_cuisine_matrix)
highest_rev = df_final_merge.groupby(['membership', 'cuisine'])['total_amount'].sum().idxmax()
highest_val = df_final_merge.groupby(['membership', 'cuisine'])['total_amount'].sum().max()
print(f"\nHighest Revenue: {highest_rev[0]} members ordering {highest_rev[1]} food (${highest_val:,.2f})")
```
